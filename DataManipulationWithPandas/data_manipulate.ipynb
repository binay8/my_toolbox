```python
import pandas as pd 
import numpy as np 
dataset = pd.read_csv("C:/Users/binay/OneDrive/Documents/GitHub/my_toolbox/Excel/store_sales_dataset.csv")
```

```python
dataset.head()
```

```python
# Calculated column 
# Get average of sales per store
dataset["AvgSales"] = dataset["Sales"]/dataset["CountSales"]
dataset.head()
```

```python
dataset["SalesTax"] = dataset["Sales"].apply(lambda x: x*0.1)
dataset.head()
```

```python
print(dataset["Sales"].mean())
print(dataset["Sales"].median())
print(dataset["Sales"].max())
print(dataset["Sales"].min())
print(dataset["Sales"].var())
print(dataset["Sales"].std())
print(dataset["Sales"].sum())
print(dataset["Sales"].quantile()) # Defaults to .5 (50% qualtile)
```

```python
# Create a function to get 25th quantile
def quantile25(field):
    return field.quantile(.25)

dataset["Sales"].agg(quantile25)
```

```python
# Use the same above function to calculate quantile of multiple columns
dataset[["Sales", "CountSales"]].agg(quantile25)
```

```python
# Create a function to get 60th Quantile
def quantile60(field):
    return field.quantile(.6)
    
```

```python
dataset[["Sales","CountSales"]].agg([quantile25, quantile60])
```

```python
# Calculate Cumulative sum
dataset["Sales"].cumsum()
```

```python
dataset["Sales"].cummin()
```

```python
dataset.sort_values("CountSales")["CountSales"].head().cumprod()
```

```python
dataset.head()
```

```python
# Dropping duplicate columns
dataset.drop_duplicates() # By default the drop_duplicates() method delets rows based on all values. Meaning if the all values are the same. 
```

```python
# To remove duplicates in a specific column, add column name to subset
dataset.drop_duplicates(subset="Year") # This keeps the firs unique entry and delets the rest
```

```python
dataset.drop_duplicates(subset=["Year","StoreType"]) # Specify multiple columns and keep unique based on that combo
```

```python
dataset.drop_duplicates(subset="Year", keep="last") # Instead of keeping the first record, specyfing keep 'last' will keep the last unique occurance instead of first by default
```

```python
dataset["Year"].value_counts() # Get count of records pre category
```

```python
dataset["StoreType"].value_counts(sort=True)
```

```python
dataset["StoreType"].value_counts(normalize = True, sort=True)
```

```python
# Summary Stat by groupe
yr_2015 = dataset["Year"]==2015
yr_2016 = dataset["Year"]==2016
yr_2017 = dataset["Year"]==2017

print(dataset[yr_2015]["Sales"].mean())
print(dataset[yr_2016]["Sales"].mean())
print(dataset[yr_2017]["Sales"].mean())
```

```python
# Grouped Summary using .groupby
dataset.groupby("Year")["Sales"].mean()
```

```python
# Grouped Summary using multiple fields and .groupby
dataset.groupby(["Year", "StoreType"])["Sales"].mean()
```

```python
dataset.groupby("StoreType")["Sales"].max()
```

```python
dataset.groupby("StoreType")["Sales"].agg([np.median, np.mean, min, max, sum])
```

```python
dataset.groupby("StoreType")["StoreType"].value_counts()
```

```python
dataset.groupby(["StoreType","StoreID"])[["Sales","CountSales"]].mean()
```

```python
dataset.pivot_table(values="Sales",index="StoreType") # By default the aggregation used in mean
```

```python
dataset.pivot_table(values="Sales", index="StoreType",aggfunc=np.median) # to specify and aggfunction other than mean add values to aggfunc=
```

```python
dataset.pivot_table(values="Sales",index="StoreType", columns = "Year")
```

```python
dataset.pivot_table(values="Sales",index="StoreType", columns = "Year", fill_value=0, margins=True )
```
